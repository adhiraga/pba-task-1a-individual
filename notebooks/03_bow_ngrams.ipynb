Feature extraction using:
- Bag of Words (BoW)
- N-Grams (Bigrams and Trigrams)

```python
import pandas as pd
import ast
from sklearn.feature_extraction.text import CountVectorizer
```

## 1. Load Preprocessed Data

```python
df = pd.read_csv('mtix_preprocessed.csv')
df['stemmed'] = df['stemmed'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else [])
df['clean_text'] = df['stemmed'].apply(lambda x: ' '.join(x))
print(f'Loaded {len(df)} rows')
df[['content', 'clean_text', 'label']].head()
```

## 2. Bag of Words (Unigram)

```python
vectorizer_bow = CountVectorizer()
X_bow = vectorizer_bow.fit_transform(df['clean_text'])

print('BoW Matrix Shape:', X_bow.shape)
print('Sample Features:', vectorizer_bow.get_feature_names_out()[:10])
```

## 3. Bigrams (N-Gram n=2)

```python
vectorizer_bigram = CountVectorizer(ngram_range=(2, 2))
X_bigram = vectorizer_bigram.fit_transform(df['clean_text'])

print('Bigram Matrix Shape:', X_bigram.shape)
print('Sample Bigrams:', vectorizer_bigram.get_feature_names_out()[:10])
```

## 4. Trigrams (N-Gram n=3)

```python
vectorizer_trigram = CountVectorizer(ngram_range=(3, 3))
X_trigram = vectorizer_trigram.fit_transform(df['clean_text'])

print('Trigram Matrix Shape:', X_trigram.shape)
print('Sample Trigrams:', vectorizer_trigram.get_feature_names_out()[:10])
```

## 5. Combined Unigram + Bigram

```python
vectorizer_ngram = CountVectorizer(ngram_range=(1, 2))
X_ngram = vectorizer_ngram.fit_transform(df['clean_text'])

print('Unigram+Bigram Matrix Shape:', X_ngram.shape)
print('Sample Features:', vectorizer_ngram.get_feature_names_out()[:10])
```
